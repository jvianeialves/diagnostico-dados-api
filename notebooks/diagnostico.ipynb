```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pymongo import MongoClient
from dotenv import load_dotenv
import os

# 1. Ler CSV - caminho correto
df = pd.read_csv('../data/dataset_exemplo.csv') 
print("CSV carregado:")
print(df.head())

# 2. Popular MongoDB
load_dotenv('../.env') # .env tá na raiz, 1 pasta acima
client = MongoClient(os.getenv('MONGO_URI'))
db = client[os.getenv('DB_NAME')] 
col = db['dispositivos'] # usando a mesma do seu código

col.delete_many({}) 
col.insert_many(df.to_dict('records'))
print(f"\nTotal no Mongo: {col.count_documents({})} documentos")

# 3. Gráfico 1: Qtd por Marca - confere com seu código
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='marca', order=df['marca'].value_counts().index)
plt.title('Quantidade por Marca'); plt.xticks(rotation=45); plt.show()

# 4. Gráfico 2: Status Ativo vs Quebrado - confere com seu código
plt.figure(figsize=(6,5))
sns.countplot(data=df, x='status')
plt.title('Status Ativo vs Quebrado'); plt.show()
```
